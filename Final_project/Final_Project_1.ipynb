```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import LabelEncoder
```

```python
string_cols = data.select_dtypes(include=['object']).columns
```

```python
# create a LabelEncoder object
le = LabelEncoder()

# encode string columns
data[string_cols] = data[string_cols].apply(le.fit_transform)
```

```python
data.head(2)
```

```python
correlation = data.corr()['TARGET']
correlation
```

```python
correlation = correlation.abs()
correlation
```

```python
sorted_feature = correlation.sort_values(ascending=False)
sorted_feature
```

```python
k_best = SelectKBest(score_func=f_classif, k=15)
```

```python
val = data.drop("TARGET", axis=1)
tar = data['TARGET']
```

```python
x_new = k_best.fit_transform(val,tar)
```

```python
selected_indices = k_best.get_support(indices=True)
```

```python
selected_features = data.columns[selected_indices]
selected_features
```
