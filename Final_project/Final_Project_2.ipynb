```python
annotation_csv = pd.read_csv(tr_annotation_path)
```

```python
annotation_csv.head()
```

```python
from PIL import Image
```

```python
for i in os.listdir(tr_data_path):
  img = Image.open(os.path.join(tr_data_path, i))
  print(img.size)
```

```python
# Image reshaping and annotation (csv to txt)

for idx, rows in annotation_csv.iterrows():
  image_name = rows[0]
  xmin = rows[1]
  ymin = rows[2]
  xmax = rows[3]
  ymax = rows[4]

  xc = (xmin + xmax) / 2 / 448
  yc = (ymin + ymax) / 2 / 448

  width = (xmax- xmin) / 448
  height = (ymax- ymin) / 448

  lab_path = "data_yolo/labels/train/"+str(image_name.split('.')[0])+ ".txt"

  with open(lab_path, "w") as file:
    file.write(f"0 {xc} {yc} {width} {height} \n")
```

```python
for idx, rows in annotation_csv.iterrows():
  image_name = rows[0]
  read_from = tr_data_path+"/"+str(image_name)
  image_path = "data_yolo/images/train/" + str(image_name)
  os.system( f'cp {read_from} {image_path}')
```

```python
from pathlib import Path
```

```python
# Env variable setup

parameters = {
    "train": str(Path("data_yolo/images/train")),
    "val": str(Path("data_yolo/images/train")),
    "nc": 1,
    "names": ['car']
}

with open("data_yolo/train_data.yaml", "w") as train_data_file:
  yaml.dump(parameters, train_data_file)
```
